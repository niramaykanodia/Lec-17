## Seattle Second Hand Car Market Analysis

```python
#Set Up
import pandas as pd
```

```python
# DO NOT DELETE: Print all rows of all dataframes
pd.set_option('display.max_rows', None)
pd.options.display.float_format = "{:,.2f}".format
```

```python
# Importing the Seattle data in
cars_df= pd.read_csv("seattle_car_fax_cleaned.csv")

# print the first 5 rows
cars_df.head(5)
```

```python
#Total rows in the data set
total_cars = len(cars_df)
print("There are ",total_cars,"used cars available for sales in seattle.")
```

```python
cars_df.isnull().sum()
```

## Uni Variate Analysis

```python
models_all_details_df = cars_df.groupby(["model_all_details"]).agg(total_occurances=("model_all_details","count"))
models_all_details_df.sort_values(["total_occurances"],ascending= False)
```

```python
# Calculating total occurences of each car in the cars_df
model_all_details_df = cars_df.groupby(["model_all_details"]).agg(total_occurences = ("model_all_details", "count"), 
                                                                  min_price = ("selling_price","min"),
                                                                  median_price = ("selling_price","median"),
                                                                  max_price = ("selling_price","max"), 
                                                                  min_miles = ("miles", "min"), 
                                                                  median_miles = ("miles", "median"), 
                                                                  max_miles = ("miles", "max")
                                                                 )
# Sorting Value
model_all_details_df.sort_values(["total_occurences"], ascending = False)
```

## Year

```python
year_df = cars_df.groupby(["year"]).agg(total_occurences = ("model_all_details", "count"), 
                                                      min_price = ("selling_price","min"),
                                                      median_price = ("selling_price","median"),
                                                      max_price = ("selling_price","max"), 
                                                      min_miles = ("miles", "min"), 
                                                      median_miles = ("miles", "median"), 
                                                      max_miles = ("miles", "max")
                                                                 )


year_df.sort_values(["total_occurences"], ascending = False)
         
```

## Brand and Varient

```python
brand_df = cars_df.groupby(["brand"]).agg(total_occurences = ("model_all_details", "count"), 
                                                      min_price = ("selling_price","min"),
                                                      median_price = ("selling_price","median"),
                                                      max_price = ("selling_price","max"), 
                                                      min_miles = ("miles", "min"), 
                                                      median_miles = ("miles", "median"), 
                                                      max_miles = ("miles", "max")
                                                                 )


brand_df.sort_values(["total_occurences"], ascending = False)
```

## Sub Model and Varient

```python
sub_model_df = cars_df.groupby(["sub_model", "brand"]).agg(total_occurences = ("model_all_details", "count"), 
                                                      min_price = ("selling_price","min"),
                                                      median_price = ("selling_price","median"),
                                                      max_price = ("selling_price","max"), 
                                                      min_miles = ("miles", "min"), 
                                                      median_miles = ("miles", "median"), 
                                                      max_miles = ("miles", "max")
                                                                 )


sub_model_df.sort_values(["total_occurences"], ascending = False)
```
